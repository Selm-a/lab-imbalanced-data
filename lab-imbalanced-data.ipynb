```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks
```

```python
df = pd.read_csv("C:/Users/Selma/Desktop/Ironhack/lab_git/unit_4_ML/week_1/lab-imbalanced-data/files_for_lab/customer_churn.csv")
df
```

```python

X = df[["SeniorCitizen", "tenure","MonthlyCharges"]]
y = df.Churn
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state =0) ## if we write 0, by default the split is 0.20
```

```python
y.value_counts()/len(y)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

print(classification_report(pred_train, y_train))
print(classification_report(pred_test, y_test))
```

```python
sm = SMOTE(k_neighbors= 3) 

X_train_SMOTE, y_train_SMOTE = sm.fit_resample(X_train, y_train)
```

```python
y_train_SMOTE.value_counts()/len(y_train_SMOTE)
```

```python
model = LogisticRegression()
model.fit(X_train_SMOTE, y_train_SMOTE)
pred_train = model.predict(X_train_SMOTE)
pred_test = model.predict(X_test)

print(classification_report(pred_train, y_train_SMOTE))
print(classification_report(pred_test, y_test))
```

```python

tl = TomekLinks()

X_train_tl, y_train_tl = tl.fit_resample(X_train, y_train)

```

```python
y_train_tl.value_counts()
```

```python
model = LogisticRegression()
model.fit(X_train_tl, y_train_tl)
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)

print(classification_report(pred_train, y_train))
print(classification_report(pred_test, y_test))
```
